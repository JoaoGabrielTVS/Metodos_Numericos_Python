```python
def escalona(A,b):
    Ab = np.c_[A,b]
    n = len(A)
    for c in range(n-1):
        for l in range(c+1,n):
            Ab[l] = Ab[l] - (Ab[l,c]/Ab[c,c])*Ab[c]
    return Ab[:,:-1], Ab[:,-1]
```

```python
print('A = \n', A)
print('\nb = ',b)
```

temos

```python
A,b = escalona(A,b)
print('A_ = \n', A)
print('\nb_ = ',b)
```

que é o mesmo resultado que obtemos!

```python
x = sla.solve(A,b)
x
```
