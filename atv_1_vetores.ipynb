```python
import numpy as np

def norm_manual(v):
    v = np.asarray(v).reshape(-1)
    return np.sqrt(np.sum(v ** 2))

v = np.array([1, 2, 3, 7, 8, 9])

print(norm_manual(v))
print(np.linalg.norm(v))
```

```python
def unit_vector(v):
    v = np.asarray(v, dtype=float)
    n = np.linalg.norm(v)
    if n == 0:
        return None
    return v / n
```

```python
def vector_with_magnitude(v, mag):
    u = unit_vector(v)
    if u is None:
        return None
    return u * mag
```

```python
import numpy as np

def row_to_col(row_vec):
    row_vec = np.asarray(row_vec)
    col = np.zeros((row_vec.shape[1], 1))
    for i in range(row_vec.shape[1]):
        col[i, 0] = row_vec[0, i]
    return col
```

```python
v = np.array([1, 2, 3, 4])

print(np.linalg.norm(v) ** 2)
print(np.dot(v, v))
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3, 1], dtype=float)
b = np.array([2, 4], dtype=float)

beta = np.dot(a, b) / np.dot(a, a)
proj = beta * a

plt.figure()
plt.plot([0, b[0]], [0, b[1]], label="b")
plt.plot([0, proj[0]], [0, proj[1]], label="proj of b on a")
plt.plot([b[0], proj[0]], [b[1], proj[1]], linestyle="--", label="residual")
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

t = np.array([4, 2], dtype=float)
r = np.array([3, 1], dtype=float)

t_parallel = r * (np.dot(t, r) / np.dot(r, r))
t_perp = t - t_parallel

print("soma:", t_parallel + t_perp)
print("ortogonalidade:", np.dot(t_perp, r))
```

```python
import numpy as np

scalars = [1, 2, -3]
vectors = [
    np.array([4, 5, 1]),
    np.array([-4, 0, -4]),
    np.array([1, 3, 2])
]

out = np.zeros(3)
for s, v in zip(scalars, vectors):
    out += s * v

print(out)
```
