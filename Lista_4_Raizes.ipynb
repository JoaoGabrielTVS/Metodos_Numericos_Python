```python
#questao 1 
b =  3
a = 1

def f(x):
    return x**3 - 3
def funcao():
    i=0
    x = 0
    global a,b
    ai =  a
    bi = b
    if (np.sign(f(a)) * np.sign(f(b))) != -1:
        return 0
    elif (np.sign(f(a)) * np.sign(f(b))) == -1:
        while(i < 1000):
            x = (ai + bi) / 2
            if(f(x) == 0):
                return x
            else:
               if np.sign(f(ai))*np.sign(f(x)) == -1:
                   bi = x
               else: 
                   ai = x
            i = i+1
    return x        
            


funcao()
```
