```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

##**Questão 1**

```python

def f(x):
  return np.cos(x)
def ponto_fixo(chute,limite, erro_aceito):
  x = f(chute)
  xn = chute
  contador = limite
  while((np.abs(x - xn))*100  >= erro_aceito or  contador==limite):

    xn = x
    x = f(xn)
    contador += 1

  return [x, xn]


ponto_fixo(1,100,1)








```

##**Questão 2**

```python
def f(x):
  return 2*np.sin((x)**(1/2))
print(ponto_fixo(0.5,100, 0.001))
```

##**Questão 3**

```python
def f(x):
  return (2*x**3 - 11.7*x**2 + 17.7*x - 5)

#parte a
x = np.arange(-2.5,2.5,1)

plt.plot(x,f(x))
plt.grid()


#parte b
b =  1
a = 0

opt.bisect(f, a, b, xtol=1e-12)
```
